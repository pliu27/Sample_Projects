# Linear regression: sklearn and statsmodels

```python
housing.dtypes
housing.info()
```

```python
sorted_housing = housing.sort_values(['total_bedrooms'], ascending = False)
sorted_housing.head()
```

```python
housing['upper'] = housing['median_house_value'].mean() + 3*housing['median_house_value'].std()
housing['lower'] = housing['median_house_value'].mean() - 3*housing['median_house_value'].std()
housing['outlier'] = (housing['median_house_value'] > housing['upper']) | (housing['median_house_value'] < housing['lower'])
housing['outlier'].unique()
```

```python
housing['median_house_value'].unique()
len(housing['median_house_value'])
len(housing['median_income'])
```

```python
## Linear regression using sklearn
from sklearn import linear_model
```

```python
x = housing[['median_income']]
y = housing[['median_house_value']]
#y.shape
lm = linear_model.LinearRegression()
model = lm.fit(x,y)
lm.coef_
lm.score(x,y) #R square of the model
predictions = lm.predict(x)
predictions[0:5]
lm.intercept_
```

```python
## Linear regression using statsmodel
import statsmodels.api as sm
x = sm.add_constant(x)
z = housing[['total_bedrooms','median_house_value']] #multiple regression
model = sm.OLS(y, x).fit()
predictions = model.predict(x)
model.summary()
```
